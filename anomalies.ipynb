```python
# Load the data for the RMI Brussels dataset

year, month, day, Tmax, Tmin = np.loadtxt("./data/data_1878_2023.txt", unpack=True)
ds = Dataset(year, month, day, Tmax, Tmin)
ds.set_all()
```

```python
# Store the index ranges for the GISTEMP reference of 1951-1980

i1 = ds.yearset.index(1951)
i2 = ds.yearset.index(1981)

refrange = ds.T_year[i1:i2] # Store avg. yearly temperatures over the GISTEMP reference of 1951-1980
Tref = np.mean(refrange) # The mean over the reference period is the reference temp.

anomalies = [temp - Tref for temp in ds.T_year]
anomalies_10yr = [np.mean(anomalies[i-4:i+6]) for i in range(4, len(anomalies)-6)] # Store anomalies as 10yr running averages
```

```python
# Load global anomalies from GISTEMP v4 and store as 10yr running averages

year_gt, anomaly_gt, smoothedanomaly_gt = np.loadtxt('./data/gistemp.txt', unpack=True)
anomaly_gt_10yr = [np.mean(anomaly_gt[i-4:i+6]) for i in range(4, len(anomaly_gt)-6)]

i1_gt = list(year_gt).index(1951)
i2_gt = list(year_gt).index(2018)
```

```python
i60 = ds.yearset.index(1960)
i60_gt = list(year_gt).index(1960)

i_end = ds.yearset.index(2005) 
i_end_gt = list(year_gt).index(2005) 

# Compute anomaly differences over the chosen range, both local and global
an_diff = anomalies_10yr[i_end] - anomalies_10yr[i60-4] # Beware of the 4 year offset for 10yr averaged anomalies
an_diff_gt = anomaly_gt_10yr[i_end_gt] - anomaly_gt_10yr[i60_gt-4]

print('LGWR = ' + f'{(an_diff / an_diff_gt):.2f}')
```

```python
# compute what GWL2.0 translates to wrt. the GISTEMP v4 reference period

start, end = np.where(year_gt == 1880)[0][0], np.where(year_gt == 1909)[0][0] # Take 1880-1909 as pre-industrial reference period
mean_anom = np.mean(anomaly_gt[start: end+1])

print(f'A GWL of 2.0° translates to {(2 + mean_anom):.2f}° relative to the 1951-1980 GISTEMP v4 reference')
```
