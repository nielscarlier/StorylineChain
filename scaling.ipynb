```python
year, month, day, Tmax, Tmin = np.loadtxt("./data/data_1878_2023.txt", unpack=True)
dataset_his = Dataset(year, month, day, Tmax, Tmin)
dataset_his.set_all()
```

```python
Tmax_adj = dataset_his.Tmax.copy()
Tmin_adj = dataset_his.Tmin.copy()


i_1880 = list(dataset_his.year).index(1880)
adj2 = dict()
i=0
for yr in dataset_his.yearset:
    adj2[yr] = local_diffGWL2[i-2] # Beware of offset (dataset starts at 1878 while anomalies start at 1880)
    i += 1

for i,yr in enumerate(year):
    if yr >= 1880:
        Tmax_adj[i] += adj2[yr]
        Tmin_adj[i] += adj2[yr]
    

dataset_his2 = Dataset(year[i_1880:], month[i_1880:], day[i_1880:], Tmax_adj[i_1880:], Tmin_adj[i_1880:])
dataset_his2.set_all()
```

```python
i_1960 = dataset_his.yearset.index(1960)
i_2017 = dataset_his.yearset.index(2017)

print('-------------------')
print('Historical extrema:')
print('-------------------')

# Cooling Degree Days (CDD) - Top 5
CDD_values = list(dataset_his.CDD.values())[i_1960:i_2017+1]
CDD_years = list(dataset_his.CDD.keys())[i_1960:i_2017+1]
CDD_top5_indices = np.argsort(CDD_values)[-5:][::-1]
CDD_top5 = [(CDD_values[i], CDD_years[i]) for i in CDD_top5_indices]

# Maximum Temperature (Tmax) - Top 5
Tmax_values = dataset_his.Tmax_year[i_1960:i_2017+1]
Tmax_years = dataset_his.yearset[i_1960:i_2017+1]
Tmax_top5_indices = np.argsort(Tmax_values)[-5:][::-1]
Tmax_top5 = [(Tmax_values[i], Tmax_years[i]) for i in Tmax_top5_indices]

# Print Results with formatting
print("Top 5 CDD (Cooling Degree Days):")
for val, yr in CDD_top5:
    print(f"  {int(yr)}: {val:.2f}")

print("Top 5 Tmax (Maximum Temperature):")
for val, yr in Tmax_top5:
    print(f"  {int(yr)}: {val:.2f}")

print('-------------------')
print('GWL2 extrema:')
print('-------------------')

#dataset_his.set_all(adjusted=True)  # Recalculate everything with the scaled GWL2.0 series

CDD_values = list(dataset_his2.CDD.values())[i_1960:i_2017+1]
CDD_years = list(dataset_his2.CDD.keys())[i_1960:i_2017+1]
CDD_top5_indices = np.argsort(CDD_values)[-5:][::-1]
CDD_top5 = [(CDD_values[i], CDD_years[i]) for i in CDD_top5_indices]

CDD_target_idx = np.argmax(CDD_values)
CDD_target = (CDD_values[CDD_target_idx], CDD_years[CDD_target_idx])

Tmax_values = dataset_his2.Tmax_year[i_1960:i_2017+1]
Tmax_years = dataset_his2.yearset[i_1960:i_2017+1]
Tmax_top5_indices = np.argsort(Tmax_values)[-5:][::-1]
Tmax_top5 = [(Tmax_values[i], Tmax_years[i]) for i in Tmax_top5_indices]

Tmax_target_idx = np.argmax(Tmax_values)
Tmax_target = (Tmax_values[Tmax_target_idx], Tmax_years[Tmax_target_idx])

print("Top 5 CDD (Cooling Degree Days):")
for val, yr in CDD_top5:
    print(f"  {int(yr)}: {val:.2f}")

print("Top 5 Tmax (Maximum Temperature):")
for val, yr in Tmax_top5:
    print(f"  {int(yr)}: {val:.2f}")
```
